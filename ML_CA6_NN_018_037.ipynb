```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers.experimental import RandomFourierFeatures
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
data = pd.read_csv("glass_identification.csv")
data.head() 
```

```python
df = data[['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Type']]
df.head()
```

```python
X = df[['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']].values
Y = df[['Type']].values
print(X[:5],Y[:5])
```

```python
X = MinMaxScaler().fit(X).transform(X.astype(float))
X[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
model = Sequential()
model.add(Dense(5,activation="relu", input_dim=9))
model.add(Dense(8, activation="sigmoid"))
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy")
model.fit(X_train,y_train,epochs=1000)
```

```python
y_hat = model.predict(X_test)
y_hat = y_hat.argmax(axis=1)
print("Accuracy Score : ",np.round(metrics.accuracy_score(y_test,y_hat)*100,2),"%")
print(confusion_matrix(y_test,y_hat))
print(classification_report(y_test,y_hat))
```

```python
model_1 = Sequential()
model_1.add(Dense(5,activation="relu", input_dim=9))
model_1.add(RandomFourierFeatures(8, kernel_initializer="gaussian"))
model_1.compile(optimizer="adam", loss="sparse_categorical_crossentropy")
model_1.fit(X_train,y_train,epochs=1000)
```

```python
y_hat = model_1.predict(X_test)
y_hat = y_hat.argmax(axis=1)
print("Accuracy Score : ",np.round(metrics.accuracy_score(y_test,y_hat)*100,2),"%")
print(confusion_matrix(y_test,y_hat))
print(classification_report(y_test,y_hat))
```
